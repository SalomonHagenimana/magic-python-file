# 3.a
The system we have is $$\frac{dx}{dt}=y$$and $$\frac{dy}{dt}=-x$$

```python
import numpy as np

def EulerMethod(a,b):
    change_in_time = 0.01 
    N = int((b-a)/change_in_time) +1
    t = np.linspace(a,b,N)
    x = np.ones(N)
    y = np.ones(N)
    x[0] = 1
    y[0] = -1

    for i in range(N-1):
        x[i+1] = x[i] + change_in_time * y[i]
        y[i+1] = y[i] - change_in_time * x[i]
        
    return t, x, y
```

```python
t, y, z = EulerMethod(0,100)
# Exact solution on euler method 
exact = lambda t: np.cos(t) - np.sin(t)
```

```python
import matplotlib.pyplot as plt
plt.plot(t,y)
plt.plot(t,exact(t))
plt.xlabel("t")
plt.ylabel("(y,exact result)")
```

# 3.b
The system we have is $$\frac{dx}{dt}=y$$and $$\frac{dy}{dt}=-x$$


```python
import numpy as np

acceleration = lambda x: -x
#dt means change in time
def VerletLeapFrogMethod(a,b, accel):
    dt = 0.01 
    N = int((b-a)/dt) + 1
    t = np.linspace(a,b,N)
    x = np.ones(N)
    y = np.ones(N)
    x[0] = 1
    y[0] = -1

    for i in range(N-1):
        x[i+1] = x[i] + (dt * y[i]) + (0.5 * dt**2 * accel(x[i]))
        y[i+1] = y[i] + (0.5 * dt * (accel(x[i+1]) + accel(x[i])))
        
    return t, x, y
```

```python
t, x, y = VerletLeapFrogMethod(0,100, acceleration)
plt.plot(t,x)
plt.plot(t,exact(t))
plt.xlabel("t")
plt.ylabel("(x,exact)")
plt.show()
```

# 3.d
$$\frac{dx}{dt}=y$$ and $$\frac{dy}{dt}=-x-\beta y$$

```python
import numpy as np

acceleration_a = lambda x: -x

def LeapFrMeWDamping(a,b, ac):
    beta = 0.2
    change_in_time = 0.01 
    N = int((b-a)/change_in_time) + 1
    t = np.linspace(a,b,N)
    x = np.zeros(N)
    y = np.zeros(N)
    x[0] = 1
    y[0] = -1

    for i in range(N-1):
        x[i+1] = x[i] + (change_in_time * y[i]) + (0.5 * change_in_time**2 * (ac(x[i]) - (beta *y[i])))
        y[i+1] = y[i] + (0.5 * change_in_time* ((ac(x[i+1])- (beta * y[i+1])) + (ac(x[i]) - (beta *y[i]))))
        
    return t, x, y
```

```python
t_1, x_1, y_1 = LeapFrMeWDamping(0,30, acceleration_a)
```

# 3.e
$$\frac{dx}{dt}=y$$ and $$\frac{dy}{dt}=-x-\beta \sin y$$

```python
acceleration1 = lambda x: -x

def LeapFrogMethodDa(p,q, ac):
    beta = 0.1
    change_in_time = 0.01 
    N = int((q-p)/change_in_time) + 1
    t = np.linspace(p,q,N)
    x = np.zeros(N)
    y = np.ones(N)
    x[0] = 1
    y[0] = -1

    for i in range(N-1):
        x[i+1] = x[i] + (change_in_time* y[i]) + (0.5 * change_in_time**2 * (ac(x[i]) - (beta *np.sin(y[i]))))
        y[i+1] = y[i] + (0.5 * change_in_time * ((ac(x[i+1])- (beta * np.sin(y[i+1]))) + (ac(x[i]) - (beta * np.sin(y[i])))))
        
    return t, x, y
```
